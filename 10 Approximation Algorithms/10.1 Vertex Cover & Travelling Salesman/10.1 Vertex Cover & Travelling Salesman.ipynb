<div id="section3" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 22px; color: #ff6600; font-weight: bold">
فروشنده دوره‌گرد
</div>
    <div id="section3.1">
    <div style="font-size: 17px; color: #ff6600; font-weight: bold">
        معرفی مسئله
</div>
    <ul>
    مسئله فروشنده دوره‌گرد 
    (Travelling Salesman Problem)
    یا به اختصار TSP
    از مسائل معروف NP
    است که بسیار مطالعه شده‌است.
    در این مسئله تعدادی شهر داریم که نمایانگر رئوس و بین هر دو شهر جاده‌ای داریم که نمایانگر یال‌ها می‌باشند طی کردن هر جاده مدت زمانی طول می‌کشد. حال یک فروشنده دوره‌گرد سعی می‌کند از شهری شروع کند، کل شهر‌ها را یکبار ببیند و سر جای اولیه‌اش برگردد. دغدغه او پیدا کردن چنین دوری است به طوریکه مدت زمانی که در جاده‌ها سپری می‌کند کمینه شود.
    <br>
    به بیان ساده‌تر مسئله فروشنده دوره‌گرد پیدا کردن کوتاهترین دور همیلتونی در یک گراف کامل بدون جهت وزن‌دار با وزن‌های مثبت است.
    <br>
    این مسئله یک مسئله NP-Hard است و نسخه تصمیم‌گیری‌اش به این صورت است:
    <br>
    گرافی وزن‌دار و یک عدد $k$ داده‌شده‌است
    . حال می‌خواهیم ببینیم آیا دور همیلتونی‌ای وجود دارد که وزنش کمتر‌ یا مساوی $k$ باشد.
        </ul>
    </div>
    </div>
</div>    

<div id="section3.2" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div style="font-size: 17px; color: #ff6600; font-weight: bold">
        راه‌حل حریصانه
    </div>
    <ul>
    حال که با مسئله NP-Hard طرفیم سعی می‌کنیم به سراغ راه‌حل‌های تقریبی برای حل این مسئله برویم.
    برای راحتی کار فعلاً فرض کنید گراف ما اقلیدسی است. یعنی هر رأس معادل نقطه‌ای روی صفحه مختصات است و یال بین دو رأس بیانگر فاصله بین آن دو است.
    <br>
    حال به سراغ حل این حالت می‌رویم. فرض کنید از رأس ۱ شروع می‌کنیم و روشی حریصانه انتخاب می‌کنیم و هر سری بین رئوسی که هنوز ندیدیم رأسی را انتخاب می‌کنیم که فاصله‌اش با یکی از رئوس مجموعه رئوس دیده‌شده کمینه باشد. یعنی یک مجموعه 
    $S$
    داریم که در ابتدا در آن ۱ قرار دارد سپس بین رئوس خارج از 
    $S$
    رأسی را انتخاب می‌کنیم که به $S$ از همه رئوس دیگر نزدیکتر است.
    یعنی وزن مینیمم یالش به این مجموعه کمینه‌است. سپس این رأس را به $S$
    اضافه می‌کنیم 
       فرض کنید در تمامی مراحل می‌خواهیم دوری را نگه داریم و بزرگ کنیم. فرض کنید پس از سه مرحله رئوس $a$ و $b$ و $c$
    به مجموعه اضافه شده‌اند. دور $C$ را برابر این سه‌تایی تعریف می‌کنیم.
    <br>
    سپس در هر مرحله که رأسی مثل $j$ قرار است به مجموعه اضافه شود،
    فرض کنید کوتاهترین یالش به $S$ یالش به رأس $i$ باشد.
    اگر $k$ رأس مجاور $i$ در دور $C$ باشد،
    یال $ik$ را حذف کرده و دو یال $ij$ و $jk$ 
    را به 
    $C$ اضافه می‌کنیم.
    <br>
    حال سراع اثبات این الگوریتم می‌رویم.
    <br>
    الگوریتم در انتخاب کردن رئوس بسیار شبیه به الگوریتم Prim در پیدا کردن درخت کمینه فراگیر عمل می‌کند.
    بنابراین مقایسه وزن کوچکترین درخت فراگیر و کوچکترین دور همیلتونی امری طبیعی است:
    <br>
    <br>
    <span style="color: red">
        لم:
        </span>
        وزن دور همیلتونی کمینه بزرگتر مساوی 
    کوچکترین درخت فراگیر است
    <br>
    اثبات چنین چیزی خیلی راحت است چرا که با حذف یک یال از دور همیلتونی یک مسیر داریم که فراگیر است و چون هر مسیر یک درخت است وزن این مسیر قاعدتاً بیشتر مساوی درخت فراگیر کمینه است در نتیجه وزن این دور بیشتر از درخت فراگیر کمینه است.
    <br>
    <br>
    حالا اگر ثابت کنیم الگوریتم داده‌شده وزنش حداکثر ۲ برابر کوچکترین درخت فراگیر است می‌توانیم به سادگی ثابت کنیم الگوریتم 
    $2-approx$
    می‌باشد
    $$ans \le 2 \times MST \le 2 \times OPT \ \to \ ans \le 2 \times OPT$$
    برای اثبات این موضوع از درستی الگوریتم Prim استفاده می‌کنیم.
    می‌دانیم رأس $ij$ رأسی از درخت فراگیر کمینه برای مجموعه $S$ است.
    اگر وزن $C$ را قبل از تغییرات $c$ بنامیم
    اکنون این وزن اینگونه تغییر کرده‌است:
    $$c' = c - w_{ik} + w_{ij} + w_{jk}$$
    اگر وزن درخت پوشای کمینه در هر مرحله $t$ باشد اکنون به این‌صورت تغییر کرده‌است:
    $$t' = t + w_{ij}$$
    نامساوی مثلثی برای یال‌ها برقرار است یعنی داریم:
    $$w_{jk} \le w_{ij} + w_{ik} \to w_{jk} - w_{ik} \le w_{ij}$$
    بنابراین:
    $$c' = c - w_{ik} + w_{jk} + w_{ij} \le c + 2w_{ij} \le 2t + 2w_{ij} \le 2t' \to c' \le 2t'$$
    در نتیجه الگوریتمی چند جمله‌ای داریم که می‌تواند با ضریب تقریب ۲ مسئله TSP را روی گراف اقلیدسی حل کند!
    به طور دقیق‌تر هر گرافی که نامساوی مثلثی به‌ازای یال‌هایش برقرار باشد این الگوریتم برایش جواب‌گو خواهد بود!
    به این گراف‌ها گراف‌های 
    <strong>متریک</strong>
    می‌گویند.
    <br>
    <br>
    <div style="color: red">
    روش راحتتر برای پیاده‌سازی الگوریتم:
        </div>
    فرض کنید به هر روشی یک درخت پوشای کمینه را پیدا‌کردیم. حالا روی این درخت dfs می‌زنیم و رئوس را به ترتیب دیدنشان کنار هم می‌گذاریم.
    می‌توان ثابت کرد این لیست اگر به ترتیب بپیماییم و در نهایت به رأس شروعمان برگردیم دور همیلتونی حاصل نیز حداکثر ۲ برابر وزن دور همیلتونی کمینه را دارد.
    </ul>
    </div>
    

<div id="section3.3" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div style="font-size: 17px; color: #ff6600; font-weight: bold">
        راه‌حلی بهتر
    </div>
    <ul>
    در ادامه به بررسی روشی می‌پردازیم که کمی تقریب بهتری روی گراف‌های متریک به ما می‌دهد.
    کلید این راه‌حل این است که سعی کنیم یک تور اویلری روی گراف پیدا‌کنیم که وزنش کم است. به صورت دقیق‌تر حداکثر
    $\frac{3}{2}$
    وزن کوچکترین درخت فراگیر است.
    <br>
    فرض کنید بازهم کوچکترین درخت فراگیر کمینه را پیدا کردیم، حالا می‌خواهیم تعدادی یال به این درخت اضافه کنیم تا درجه همه‌ رئوس زوج شود، در این‌صورت می‌دانیم تور اویلری‌ای وجود دارد که شامل کل رئوس بشود. برای اینکار رئوس درجه فرد را در مجموعه‌ای مثل $O$ در نظر می‌گیریم.
    حال اگر تطابقی از این رئوس پیدا کنیم و آن‌را به مجموعه یال‌ها اضافه کنیم هر رأسی که قبلا درجه‌اش فرد بوده زوج می‌شود. توجه کنید ممکن است یک یال دوبار تکرار شود در ادامه توضیح می‌دهیم مشکلی بوجود نخواهد آمد.
    <br>
    حالا نکته‌ای که وجود دارد این است که این 
    $\frac{|O|}{2}$
    یال اضافه را طوری انتخاب کنیم که مجموع وزن یال‌ها کمینه شود. برای این‌کار می‌توانیم از الگوریتم کوچکترین تطابق وزن‌دار استفاده نماییم.
    <br>
    حال به لم کلیدی زیر توجه کنید:
    <br>
    <br>
    <span style="color: red;">لم:</span>
    اگر کوچکترین دور همیلتونی کمینه در یک گراف متریک 
    $OPT$
    باشد آن وقت کوچکترین تطابق برای مجموعه 
    $O$
    اندازه‌اش حداکثر
    $\frac{OPT}{2}$
    است.
    <br>
     دور همیلتونی کمینه را در نظر بگیرید و رئوس $O$ را روی آن در نظر بگیرید. اگر این رئوس را به ترتیب شماره گذاری کنیم یکی از دو تطابق 
    $$\{o_1, o_2\}, \{o_3, o_4\}, ..., \{o_{n-1}, o_n\}$$
    یا
    $$\{o_2, o_3\}, \{o_4, o_5\}, ..., \{o_n, o_1\}$$
    وزنش حداکثر 
    $\frac{OPT}{2}$
    است (چرا؟)
<br>
    <br>
    بنابراین حالا که لم ثابت شد می‌توانیم تور اویلری‌ای بیابیم که مجموع وزن یال‌هایش حداکثر 
    $\frac{3}{2}OPT$
    است.
    حالا کافیست در تور اویلری حرکت کنیم و اگر بین رأس $u$ و $v$
    تعدادی رأس تکراری مثل $x_1, x_2,..., x_k$
    وجود داشت آن‌ها را نبینیم و به جای آن مستقیم از $u$ به $v$ برویم.
    به‌دلیل متریک بودن گراف داریم:
    $w_{uv} \le w_{u,x_1} + w_{x_1, x_2} + ... + w_{x_k, v}$
    بنابراین دوری که با انجام این‌کار می‌بینیم وزنی کمتر مساوی تور اویلری دارد که خود کمتر مساوی 
    $\frac{3}{2}OPT$
    است.

<div style="font-size: 22px; color: #ff6600; font-weight: bold">
سختی تقریب
</div>
    برای تحلیل سختی یک الگوریتم تقریبی با ضریب تقریب 
    $\alpha$
    راه‌های مختلفی وجود دارد. یکی از مرسوم ترین راه‌ها کاهش دادن مسئله به مسئله 
    $NP-Complete$
    است.
    اینطور که می‌گوییم اگر فرض کنیم $P \neq NP$
    در این‌صورت الگوریتمی 
    $\alpha-approx$
    برای مسئله X نداریم.
       <br>
    برای مثال بر می‌گردیم به مسئله فروشنده دوره‌گرد! در حالت متریک الگوریتم‌های تقریبی خوبی به‌دست آوردیم. اما حالا می‌خواهیم ببینیم آیا برای حالت کلی می‌توان برای مثال الگوریتم تقریبی 
    $2-approx$
    ارائه کرد؟
    <br>
    <br>
    <span style="color: red">قضیه:</span>
    اگر $P \neq NP$ در این‌صورت الگوریتمی چند‌جمله‌ای 
    $2-approx$ برای مسئله TSP وجود ندارد.
    <br>
    اثبات: فرض می‌کنیم الگوریتم چند
    جمله‌ای و 
    $2-approx$
    برای این‌کار داریم. با استفاده از این الگوریتم راه‌حلی چند جمله‌ای برای مسئله 
    NP-Complete
    وجود داشتن دور همیلتونی ارائه می‌کنیم.
    به این‌صورت که یال‌های داخل گراف را با وزن ۱ وزن‌دار می‌کنیم و یال‌هاییکه ظاهر نشده‌اند با $n+2$ وزن‌دار می‌کنیم.
    اگر الگوریتم تقریبی را اجرا کنیم و عددی 
    کمتر از $2n+1$ 
    حاصل شد یعنی این گراف دور همیلتونی دارد.
    و در غیر اینصورت فاقد دور همیلتونی است. (توجه کنید این وزن‌گذاری متریک نیست!)
    <br>
    <br>
    می‌توانیم از این هم فراتر برویم فرض کنید الگوریتمی 
    $\alpha-approx$
    برای این مسئله داشته‌باشیم. در این‌صورت اگر وزن یال‌های گراف را برابر ۱ و وزن‌های دیگر را برابر
    $(\alpha - 1) \times n + 1$
    بگذاریم می‌توانیم وجود یا عدم وجود دور همیلتونی را تشخیص دهیم بنابراین. به ازای 
    <strong>
    هر عدد ثابت 
    </strong>
    $\alpha \ge 1$
    <strong>
    هیچ الگوریتم تقریبی چند جمله‌ای 
    </strong>
    $\alpha-approx$
    برای مسئله TSP نداریم!
      <br>
    <br>
    <span style="color: red">تمرین:</span>
    مسئله $k-centre$ مسئله‌ای است که در آن
    تعدادی نقطه در صفحه مختصات داریم. می‌خواهیم این نقاط را در $k$ گروه تقسیم کنیم و برای هر گروه یک نقطه تحت عنوان 
    مرکز بکشیم. وزن هر دسته را برابر ماکسیمم فاصله نقاط آن دسته از مرکزش می‌گذاریم. می‌خواهیم تقسیم بندی و مرکز‌گذاری‌ای را بیابیم به‌طوریکه ماکسیمم وزن مینیمم شود.
    <br>
    الف) سعی کنید الگوریتمی 
    چند‌جمله‌ای
    $2-approx$
     برای این‌کار 
    ارائه دهید.
    <br>
    ب) ثابت کنید هیچ الگوریتم چند جمله‌ای 
    $\alpha-approx, \alpha < 2$
    برای این‌کار وجود ندارد مگر این‌که
    $P = NP$
    <br>
    <br>
    <span style="color: blue">راهنمایی:</span>
    برای بخش ب) سعی کنید با مسئله کوچکترین مجموعه غالب تناظر دهید. مجموعه غالب در یک گراف برابر مجموعه رأسی است که هر رأس در گراف یا داخل مجموعه غالب است یا اینکه به یکی از اعضای مجموعه غالب یال دارد.
</div>
